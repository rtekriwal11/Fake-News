```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracy and recall values for supervised and semi-supervised learning
accuracy_supervised = 89.15
recall_supervised=90.8

accuracy_semi_1st_set = 89.6
recall_semi_1st_set = 90.25

accuracy_semi_2nd_set = 89.55
recall_semi_2nd_set = 90.83

accuracy_semi_3rd_set = 89.25
recall_semi_3rd_set = 90.31

# Create a bar graph
labels = ['Supervised', 'Semi-Supervised (1st Set)', 'Semi-Supervised (2nd Set)', 'Semi-Supervised (3rd Set)']
accuracy_values = [accuracy_supervised, accuracy_semi_1st_set, accuracy_semi_2nd_set, accuracy_semi_3rd_set]
recall_values = [recall_supervised, recall_semi_1st_set, recall_semi_2nd_set, recall_semi_3rd_set]

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting accuracy values
ax1.bar(labels, accuracy_values, color='blue', alpha=0.7, label='Accuracy')

# Plotting recall values on the secondary axis
ax2 = ax1.twinx()
ax2.bar(labels, recall_values, color='orange', alpha=0.7, label='Recall')

ax1.set_ylabel('Accuracy')
ax2.set_ylabel('Recall')

plt.title('Comparison of Accuracy and Recall in Supervised and Semi-Supervised Learning')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracy and recall values for supervised and semi-supervised learning
accuracy_supervised = 89.15

accuracy_semi_1st_set = 89.6
recall_semi_1st_set = 90.25

accuracy_semi_2nd_set = 89.55
recall_semi_2nd_set = 90.83

accuracy_semi_3rd_set = 89.25
recall_semi_3rd_set = 90.31

# Create a bar graph
labels = ['Supervised', 'Semi-Supervised (1st Set)', 'Semi-Supervised (2nd Set)', 'Semi-Supervised (3rd Set)']
accuracy_values = [accuracy_supervised, accuracy_semi_1st_set, accuracy_semi_2nd_set, accuracy_semi_3rd_set]
recall_values = [np.nan, recall_semi_1st_set, recall_semi_2nd_set, recall_semi_3rd_set]

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting accuracy values
ax1.bar(labels, accuracy_values, color='blue', alpha=0.7, label='Accuracy (Supervised vs. Semi-Supervised)')

# Plotting recall values on the secondary axis
ax2 = ax1.twinx()
ax2.bar(labels, recall_values, color='orange', alpha=0.7, label='Recall (Semi-Supervised)')

ax1.set_ylabel('Accuracy')
ax2.set_ylabel('Recall')

plt.title('Comparison of Accuracy and Recall in Supervised and Semi-Supervised Learning')
plt.legend(loc='upper left', bbox_to_anchor=(0.7, 0.9))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subsets = ['Subset 1', 'Subset 2', 'Subset 3']
accuracy_supervised = 89.15
recall_supervised = 90.8
accuracy_semi_supervised = [89.6, 89.55, 89.25]
recall_semi_supervised = [90.25, 90.83, 90.31]

# Plotting
plt.figure(figsize=(10, 6))

# Plot supervised accuracy and recall
plt.plot(subsets, [accuracy_supervised] * 3, label='Supervised Accuracy', marker='o', linestyle='--', color='g')
plt.plot(subsets, [recall_supervised] * 3, label='Supervised Recall', marker='o', linestyle='--', color='c')

# Plot semi-supervised accuracy and recall
plt.plot(subsets, accuracy_semi_supervised, label='Semi-Supervised Accuracy', marker='o', linestyle='-', color='b')
plt.plot(subsets, recall_semi_supervised, label='Semi-Supervised Recall', marker='o', linestyle='-', color='r')

# Adding labels and title
plt.xlabel('Subsets')
plt.ylabel('Percentage (%)')
plt.title('Accuracy and Recall Comparison')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
